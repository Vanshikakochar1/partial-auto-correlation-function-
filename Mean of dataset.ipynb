```python
import pandas as pd
import numpy as np

# Load dataset
data = pd.read_csv('station_belem.csv')

```

```python
# Basic Data Exploration
# Check first 5 rows of data
print(data.head())
```

```python
# Check data types, number of records, and missing values
print(data.info())
```

```python
# Descriptive statistics of numerical columns
print(data.describe())
```

```python
# Compute correlation matrix
corr_matrix = data.corr()

# Display correlation matrix
print(corr_matrix)
```

```python
df = pd.DataFrame(data)
# Exclude column 'YEAR' from mean calculation
cols_to_mean = df.columns.drop('YEAR')
```

```python
# Calculate mean by row, excluding column 'YEAR'
row_means = df[cols_to_mean].mean(axis=1)

# Display row means
print(row_means)
```

```python
# Add new column with row means to DataFrame
df['Row Means'] = row_means

# Display updated DataFrame
print(df)
```
